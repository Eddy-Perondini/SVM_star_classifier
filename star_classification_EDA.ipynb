```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
import kagglehub
import plotly.express as px
import os 
sns.set_style("dark")
```

```python
path = kagglehub.dataset_download("deepu1109/star-dataset")

dados = pd.read_csv(f'{path}/6 class csv.csv')
```

```python
dados.head()
```

```python
#Gerando algumas visualizações

plt.hist(x = dados['Temperature (K)'], bins = 20)
plt.xlabel('Temperatura (K)')
plt.grid(True)
plt.show()
```

```python
plt.hist(x = dados['Absolute magnitude(Mv)'], bins = 20)
plt.xlabel('Mv')
plt.grid(True)
plt.show()
```

```python
plt.hist(x = dados['Radius(R/Ro)'], bins = 20)
plt.xlabel('[R/Ro]')
plt.grid(True)
plt.show()
```

```python
#HR Diagram 

g = sns.scatterplot(x = dados['Temperature (K)'], y = dados['Absolute magnitude(Mv)'], hue = dados['Star type'], palette = 'inferno')

g.invert_yaxis()
g.invert_xaxis()

plt.xlabel(r'$[K]$')
plt.ylabel(r'$[M_{V}]$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Calculando algumas estatísticas 

print(f'Média da Temperatura: {dados["Temperature (K)"].mean()}')
print(f'Desvio Padrão da Temperatura: {dados["Temperature (K)"].std()}')

print("-------------------------")

print(f'Média da Luminosidade: {dados["Luminosity(L/Lo)"].mean()}')
print(f'Desvio Padrão da Luminosidade: {dados["Luminosity(L/Lo)"].std()}')

print("-------------------------")

print(f'Média do Raio: {dados["Radius(R/Ro)"].mean()}')
print(f'Desvio Padrão do Raio: {dados["Radius(R/Ro)"].std()}')

print("-------------------------")

print(f'Média da Magnitude Absoluta: {dados["Absolute magnitude(Mv)"].mean()}')
print(f'Desvio Padrão da Magnitude Absoluta: {dados["Absolute magnitude(Mv)"].std()}')
```

```python
#Contando a quantidade de cores de estrelas 

print(f'Quantidade de cores de estrelas: {dados["Star color"].nunique()}')
print(dados["Star color"].value_counts())
```

```python
#Analisando a classe espectral 

print(f'Quantidade de classes espectrais: {dados["Spectral Class"].nunique()}')
print(dados["Spectral Class"].value_counts())

sns.countplot(x = dados['Spectral Class'], palette = 'inferno')
plt.xlabel('Classe Espectral')
plt.ylabel('Qtd.')
plt.grid(True)
plt.show()
```

```python
fig = px.parallel_categories(dados, dimensions=['Star color', 'Spectral Class', 'Star type'], color = 'Star type', color_continuous_scale = 'inferno')
fig.show()
```

```python
#Analisando a qtd de classes de estrelas 

sns.countplot(x = dados['Star type'], hue = dados['Star type'], palette = 'inferno')
plt.xlabel('Tipo de Estrela')
plt.ylabel('Qtd.')
plt.grid(True)
plt.legend()
plt.show()
```
